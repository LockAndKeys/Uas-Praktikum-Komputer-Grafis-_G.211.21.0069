Proses Menganalisis Data Stock Csv dengan mengimport ke dalam library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

Menampilkan Data Stock

```python
df = pd.read_csv('Data stock.csv')
df
```

Menampilkan setiap atribut Purchase Price, Current Price, dan Gain/Loss dengan menggunakan grafik histogram

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
sns.histplot(df['Purchase Price ($)'], kde=True, bins=20,color="purple", edgecolor='k')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price')

plt.subplot(1, 3, 2)
sns.histplot(df['Current Price ($)'], kde=True, bins=20, color="green", edgecolor='k')
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Histogram of Current Price')

plt.subplot(1, 3, 3)
sns.histplot(df['Gain/Loss ($)'], kde=True, bins=20, color="maroon", edgecolor='k')
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss')

plt.tight_layout(pad=2)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
labels = ['Purchase Price', 'Current Price', 'Gain/Loss']
heights = [len(df['Purchase Price ($)']), len(df['Current Price ($)']), len(df['Gain/Loss ($)'])]
colors = ['purple', 'lime', 'maroon']
plt.bar(labels, heights, color=colors)
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Distribution of Data')

plt.show()
```

```python
plt.figure(figsize=(6, 6))
labels = ['Purchase Price', 'Current Price', 'Gain/Loss']
sizes = [len(df['Purchase Price ($)']), len(df['Current Price ($)']), len(df['Gain/Loss ($)'])]
colors = ['purple', 'limegreen', 'maroon']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Data')

plt.show()
```

Menampilkan 5 Data Tertinggi yang dilihat dari Dividend/Shares (Saham)


```python
data = df.sort_values(by="Dividend/Shares", ascending=False)
top_5 = data.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x="Dividend/Shares", y="Stock Name", data=top_5)  # Menukar sumbu x dan y
plt.title("Top 5 Stocks with Highest Dividend / Shares")
plt.xlabel("Dividend/Shares")
plt.ylabel("Stock Name")
plt.show()
```

```python
data = df.sort_values(by="Dividend/Shares", ascending=False)
top_5 = data.head(5)

plt.figure(figsize=(6, 10))
plt.plot(top_5["Dividend/Shares"], range(1, len(top_5)+1), marker='o', color='maroon', linestyle='-')
plt.yticks(range(1, len(top_5)+1), top_5["Stock Name"])
plt.title("Top 5 Stocks with Highest Dividend / Shares")
plt.xlabel("Dividend/Shares")
plt.ylabel("Stock Name")
plt.gca().invert_yaxis()
plt.show()
```

Menampilkan 5 Data dengan Nilai tertinggi yang dilihat dari Nilai Pasar

```python
data = df.sort_values(by="Market Value ($)", ascending=False)
top_5 = data.head(5)

plt.figure(figsize=(9, 6))
sns.barplot(x="Market Value ($)", y="Stock Name", data=top_5)
plt.title("Top 5 Stocks with Highest Market Value ($)")
plt.xlabel("Market Value ($)")
plt.ylabel("Stock Name")
plt.show()
```

```python
data = df.sort_values(by="Market Value ($)", ascending=False)
top_5 = data.head(5)

plt.figure(figsize=(11, 4))
plt.plot(top_5["Stock Name"], top_5["Market Value ($)"], marker='o', color='purple', linestyle='-', linewidth=2)
plt.title("Top 5 Stocks with Highest Market Value ($)")
plt.xlabel("Stock Name")
plt.ylabel("Market Value ($)")
plt.xticks(rotation=45)
plt.show()
```

```python
data = df.sort_values(by="Annual Yield (%)", ascending=False)
top_7 = data.head(7)

plt.figure(figsize=(9, 8))
sns.barplot(x="Annual Yield (%)", y="Stock Name", data=top_7)
plt.title("Top 7 Stocks with Highest Annual Yield (%)")
plt.xlabel("Annual Yield (%)")
plt.ylabel("Stock Name")
plt.show()
```

```python
data = df.sort_values(by="Annual Yield (%)", ascending=False)
top_7 = data.head(7)

plt.figure(figsize=(11, 4))
plt.plot(top_7["Stock Name"], top_7["Annual Yield (%)"], marker='o', color='green', linestyle='-')
plt.title("Top 7 Stocks with Highest Annual Yield (%)")
plt.xlabel("Stock Name")
plt.ylabel("Annual Yield (%)")
plt.xticks(rotation=45)
plt.show()
```
